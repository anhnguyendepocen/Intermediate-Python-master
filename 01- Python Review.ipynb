### Manufacturing the truth:
- `<`  greater than (`<=` less than or equal too)

- `>` greater than (`>=` greater than or equal too)

- == is equal too (can also use `is`)
- `!=` not equal to (also use `is not`
- `in` asks if an element is in a collection (could be tuple, list , or a string)

````python
# Manufacture the truth with logic operators

print(3<5)
print('freedom' is 'free')
print('freedom' is not 'free')
print('i' in 'team')
````
